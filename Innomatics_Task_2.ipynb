1: Unique Words in a Sentence


```python


text = input("Enter sentence: ")

words = text.split()      # used to spilt the words int he text data
unique = set(words)     # Convert it to set so that duplicate elements will be removed and present the unique words

print("Unique words count:", len(unique))
print("Unique words:", unique)
```

2: Highest Salary from Employee Data


```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

high_employee = max(employees, key=employees.get) #stores the employee name with the highest salary
high_salary = employees[high_employee]            # stores the highest salary of the respective employeee

print("Employee with the Highest Salary:", high_employee, "-", high_salary)
```

3: Find Maximum and Minimum Values


```python
values = [45, 22, 89, 10, 66]

max_val = values[0]
min_val = values[0]

for num in values:         #simple for loop logic for finding the maximum and minimum values from the list
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num


print("List:", values)
print("Max:", max_val)
print("Min:", min_val)
```

4: Count Products Above a Price Threshold


```python
prices = [450, 1200, 899, 1500, 300]

count_product = 0

for p in prices:
    if p > 1000:
        count_product += 1

print("Products above 1000:", count_product)
```

5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]

present=0
for p in attendance:
    if p == "P":
        present += 1

total = len(attendance)

percentage = (present / total) * 100

print("Attendance Percentage:", percentage)
```

7: Count Character Frequency


```python
text = "python"

char_count = {}

for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

print(char_count)
```

10: Calculate Average Marks

```python
marks = [70, 80, 75, 65]
sum=0
for i in marks:
  sum+=i
count = len(marks)

average = sum / count

print("Average Marks:", average)
```
