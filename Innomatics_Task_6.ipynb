```python
# Problem 1: Smart Parking Lot Management System
def smart_parking_system(capacity, vehicle_logs):
    """
    Function to manage smart parking lot.
    Parameters:
        capacity (int): Maximum parking capacity
        vehicle_logs (list): List containing 'IN' and 'OUT' entries
    Returns:
        None (prints parking status)
    """

    current_vehicles = 0
    peak_usage = 0

    # Loop through vehicle logs
    for log in vehicle_logs:
        if log == "IN":
            current_vehicles += 1
        elif log == "OUT" and current_vehicles > 0:
            current_vehicles -= 1

        # Track peak usage
        if current_vehicles > peak_usage:
            peak_usage = current_vehicles

    print("Currently Parked Vehicles:", current_vehicles)
    print("Peak Parking Usage:", peak_usage)

    # Capacity check
    if current_vehicles > capacity:
        print("Parking Status: Full - Capacity Exceeded!")
    else:
        print("Parking Status: Available")


# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, vehicle_logs)
```

```python
# Problem 2: Online Food Delivery Time Estimator
def estimate_delivery_time(distance, traffic_level, weather):
    """
    Estimates delivery time based on distance, traffic and weather.
    """

    # Base time assumption: 5 minutes per km
    base_time = distance * 5

    # Traffic impact
    if traffic_level.lower() == "high":
        base_time += 15
    elif traffic_level.lower() == "medium":
        base_time += 10
    else:
        base_time += 5

    # Weather impact
    if weather.lower() == "rainy":
        base_time += 10
    elif weather.lower() == "stormy":
        base_time += 20

    print("Estimated Delivery Time:", base_time, "minutes")


# Sample Input
estimate_delivery_time(8, "High", "Rainy")
```

```python
# Problem 3: Movie Theatre Seat Occupancy Analyzer
def theatre_occupancy(total_seats, booked_seats):
    """
    Analyzes theatre seat occupancy.
    """

    booked_count = len(booked_seats)

    # Calculate occupancy percentage
    occupancy_percentage = (booked_count / total_seats) * 100

    print("Occupancy:", round(occupancy_percentage, 2), "%")

    # Show status conditions
    if occupancy_percentage == 100:
        print("Show Status: Housefull")
    elif occupancy_percentage >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Consider Opening Additional Show")
    else:
        print("Show Status: Seats Available")


# Sample Input
theatre_occupancy(200, [1]*150)
```

```python
# Problem 4: Cloud Server Load Classification System
def classify_server_load(cpu_readings):
    """
    Classifies server load based on average CPU usage.
    """

    # Calculate average CPU
    average_cpu = sum(cpu_readings) / len(cpu_readings)

    print("Average CPU Load:", round(average_cpu, 2), "%")

    # Classification logic
    if average_cpu < 50:
        print("Server Status: Normal")
    elif 50 <= average_cpu <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
classify_server_load([45, 60, 70, 85, 90])
```

```python
# Problem 5: Smart Classroom Resource Usage Monitor
def classroom_resource_monitor(resource_usage):
    """
    Identifies overused classroom resources.
    """

    overused_resources = []

    # Threshold for overuse (example: > 8 hours)
    threshold = 8

    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused_resources.append(resource)

    if overused_resources:
        print("Overused Resources:", ", ".join(overused_resources))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_resource_monitor(usage)
```

```python
# Problem 6: Online Event Registration Capacity Controller
def event_registration_controller(capacity, registrations):
    """
    Manages event registrations and prevents overbooking.
    """

    if registrations <= capacity:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Sample Input
event_registration_controller(100, 105)
```
