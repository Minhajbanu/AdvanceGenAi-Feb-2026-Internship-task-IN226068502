```python
# Problem 1: Mobile Recharge Validation System

def validate_recharge():
    valid = [199, 299, 399, 599]

    while True:  # Retry loop
        amount = int(input("Enter recharge amount: "))

        # Condition 1: Minimum recharge amount
        if amount < 50:
            print("Recharge amount must be at least ₹50")
            continue

        # Condition 2: Check valid plans
        if amount in valid:
            print("Recharge Successful!")
            break
        else:
            print("Invalid Plan. Try again.")
```

```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(products):
    for product, stock in products.items():  # Loop through dictionary
        if stock < 15:
            print(f"{product} → Reorder Alert ")
        else:
            print(f"{product} → Stock OK ")


# Example dictionary
inventory = {
    "Soap": 10,
    "Shampoo": 20,
    "Toothpaste": 5,
    "Oil": 30
}

check_inventory(inventory)
```

```python
# Problem 3: Student Result Processing System

def calculate_result(marks):
    total = 0

    # Loop to calculate total marks
    for mark in marks:
        total += mark

    average = total / len(marks)

    print(f"Average Marks: {average}")

    if average >= 50:
        return "Pass "
    else:
        return "Fail "


# Example
marks_list = [60, 55, 70, 40]
result = calculate_result(marks_list)
print("Result:", result)
```

```python
# Problem 4: Cab Fare Estimator

def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance * per_km_rate)

    # Peak hour condition
    if peak.lower() == "yes":
        fare += fare * 0.25  # 25% extra

    return fare


while True:  # Retry loop
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(distance, peak)
    print(f"Total Fare: ₹{total_fare}")

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

```python
# Problem 5: Attendance Eligibility Checker

def check_attendance(attendance_list):
    total_days = len(attendance_list)
    present_days = 0

    # Count present days
    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / total_days) * 100
    print(f"Attendance Percentage: {percentage}%")

    if percentage >= 75:
        return "Eligible "
    else:
        return "Not Eligible "


# Example
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
status = check_attendance(attendance)
print("Status:", status)
```

```python
# Problem 6: Password Strength Checker

def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return " Weak Password (Minimum 8 characters required)"

    # Loop through characters
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return " Strong Password"
    else:
        return "Weak Password (Must contain digit & special character)"


# Example
pwd = input("Enter password: ")
print(check_password(pwd))
```
