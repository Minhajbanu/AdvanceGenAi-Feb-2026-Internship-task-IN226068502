2. Pass / Fail Analyzer

Concept Used: Loop, Counter variables, Condition checking

```python
marks=[45, 78, 90, 33, 60]
pass_students=0
fail_students=0
for i in marks:
    if i>=50:
        pass_students=pass_students+1
    else:
        fail_students=fail_students+1
```

```python
print("total students passed the exam: ",pass_students);
print("total students failed in the exam: ",fail_students);
```

3. Simple Data Cleaner

Concept Used:

.strip() → removes spaces

.lower() → converts to lowercase


```python
names = [" Alice ", "bob", " CHARLIE "]
```

```python
result = []
for i in names:
  removed_spaces=i.strip();
  lower_case=removed_spaces.lower();
  result.append(lower_case);
```

```python
print("Cleaned Data:", result)
```

4. Message Length Analyzer

Concept Used: len() function, Conditional flagging

```python
messages = ["Hi", "Welcome to the platform", "OK"]
```

```python
for i in messages:
  length=len(i);
  print(f"Word: '{i}' ")
  print(f"Length: ",length);
  if length>10:
    print("long Message")
  else:
    print("short Message")
```

5. Error Message Detector

Concept Used:
Loop method

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
```

```python
error=0
for i in logs:
  if i=="ERROR":
    error=error+1
print("The number of error message is : ",error);
```
